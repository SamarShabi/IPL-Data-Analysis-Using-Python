```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("matches.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data['player_of_match'].value_counts()[0:10]
```

```python
list(data['player_of_match'].value_counts()[0:10].keys())
```

```python
plt.figure(figsize=[12,4])
plt.bar(list(data["player_of_match"].value_counts()[0:10].keys()), list(data["player_of_match"].value_counts()[0:10]), color='yellow')
plt.show()
```

```python
data['result'].value_counts()
```

```python
plt.figure(figsize=[4,4])
plt.bar(list(data['result'].value_counts().keys()), list(data['result'].value_counts()),color=['green','blue','red'])
plt.show()
```

```python
batting_first = data[data['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=[4,4])
plt.hist(batting_first['win_by_runs'], color='red')
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('Matches')
plt.show()
```

```python
data.head()
```

```python
data['winner'].value_counts()[0:3]
```

```python
plt.figure(figsize=[8,4])
plt.bar(list(data['winner'].value_counts()[0:3].keys()),list(data['winner'].value_counts()[0:3]), color=['blue','yellow','red'])
plt.show()
```

```python
plt.figure(figsize=[8,4])
plt.pie(list(batting_first['winner'].value_counts()[0:5]), labels= list(batting_first['winner'].value_counts()[0:5].keys()), autopct='%0.1f%%')
plt.show()
```

```python
field_first = data[data['win_by_wickets']!=0]
field_first.head()
```

```python
plt.figure(figsize=[4,4])
plt.pie(list(field_first['winner'].value_counts()[0:5]), labels=list(field_first['winner'].value_counts()[0:5].keys()), autopct='%0.1f%%')
plt.show()
```

```python
plt.figure(figsize=[4,4])
plt.hist(field_first['win_by_wickets'], bins=20)
plt.xlabel('wickets')
plt.show()
```
